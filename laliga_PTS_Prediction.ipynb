```python
import pandas as pd
```

```python
df_home=pd.read_excel("C:\pandas\Machine Learning\Kaggle\laliga2020.xlsx",sheet_name="Sheet1")
df_home
```

```python
df_away=pd.read_excel("C:\pandas\Machine Learning\Kaggle\laliga2020.xlsx",sheet_name="Sheet2")
df_away
```

```python
check_MP=df_home['MP']+df_away['MP']
check_W=df_home['W']+df_away['W']
check_D=df_home['D']+df_away['D']
check_L=df_home['L']+df_away['L']
check_GF=df_home['GF']+df_away['GF']
check_GA=df_home['GA']+df_away['GA']
check_GDiff=df_home['GDiff']+df_away['GDiff']
check_Pts=df_home['Pts']+df_away['Pts']
```

```python
final_laliga_table=df_home
final_laliga_table['MP']=check_MP
final_laliga_table['W']=check_W
final_laliga_table['D']=check_D
final_laliga_table['L']=check_L
final_laliga_table['GF']=check_GF
final_laliga_table['GA']=check_GA
final_laliga_table['GDiff']=check_GDiff
final_laliga_table['Pts']=check_Pts
final_laliga_table
```

```python

x=final_laliga_table[['MP','W','D','L','GDiff']]
x
```

```python
y=final_laliga_table['Pts']
y
```

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(x,y,test_size=0.1)
```

```python
reg=linear_model.LinearRegression()
reg.fit(X_train,y_train)
```

```python
reg.predict([[27,11,5,11,6]])
```

```python
reg.score(X_test,y_test)
```
